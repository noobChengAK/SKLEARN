```python
filepath = 'C:\\Users\\Tang\\Desktop\\E.Care\\food_portion.xlsx'
df_remove_guideline = pd.read_excel(filepath)
```

```python
df_remove_guideline = df_remove_guideline.astype(str)
#将所有转化为str格式
```

```python
df_remove_guideline
```

```python
row = df_remove_guideline[df_remove_guideline['portion_description'].str.contains('Guideline')]
#找到portion_description列中含有‘Guideline’的全部行
```

```python
row
```

```python
test1 = list(row.portion_description)
test1
```

```python
test2 = list(df_remove_guideline.portion_description)
test2
```

```python
ret = list(set(test2) ^ set(test1))
ret
#使用列表求差集的方法可将portion_description列中含有Guideline元素去除掉
```

```python
result = df_remove_guideline[df_remove_guideline.portion_description.isin(ret)]
result
```

```python
row2 = result[result['portion_description'].str.contains('Quantity')]
#找到portion_description列中含有‘Quantity’的全部行
test3 = list(row2.portion_description)
test4 = list(result.portion_description)
ret2 = list(set(test4) ^ set(test3))
result2 = result[result.portion_description.isin(ret2)]
```

```python
#将string格式转回dataFrame
from io import StringIO
#使用pickle转成2进制再转回
import pickle
df_string = pickle.dumps(result2)
del result2
df = pickle.loads(df_string)
```
